```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import copy
import warnings
from IPython.display import display
%matplotlib inline
```

```python
meta_df = pd.read_csv('data/big/ACS_16_5YR_B02001_metadata.csv',
                      encoding='ISO-8859-1', header=None)
df = pd.read_csv('data/big/ACS_16_5YR_B02001_with_ann.csv', encoding='ISO-8859-1', 
                 header=0, skiprows=[0], names=meta_df[1])
df['Id'] = df['Id'].str.slice(-5)
df.head(5)
```

```python
tmp = [x.startswith('Margin') for x in df.columns]
df.drop(df.columns[tmp], axis=1, inplace=True)
df.drop(['Id2'], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df.columns = [x.replace('Estimate; Total: - ','') for x in df.columns]
df.columns = [x.replace(':','') for x in df.columns]
df.columns = [x.replace(';','') for x in df.columns]
df.columns = [x.replace(' ','_') for x in df.columns]
df.head(5)
```

```python
# df['Id'] = df['Id'].str.slice(-5)
for col in df.columns[3:]:
    df[col] = df[col] / df['Estimate_Total']
```

```python
df.to_csv('data/ACS_16_pop_demos.csv', index=False)
```
